These are your tasks for this mini project:

1. Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD).
2. Convert the returned JSON object into a Python dictionary.
3. Calculate what the highest and lowest opening prices were for the stock in this period.
4. What was the largest change in any one day (based on High and Low price)?
5. What was the largest change between any two days (based on Closing Price)?
6. What was the average daily trading volume during this year?
7. (Optional) What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)

```python
# Get the metadata of Zeiss Stock Data
m_url = "https://www.quandl.com/api/v3/datasets/FSE/AFX_X/metadata.json?api_key="
m_res = requests.get(m_url + API_KEY)
m_res.json()['dataset']['column_names']
```

```python
# Get time-series data of Zeiss Stock 
url = "https://www.quandl.com/api/v3/datasets/FSE/AFX_X.json?&start_date=2017-01-01&end_date=2017-12-31&order=asc&api_key="
res = requests.get(url + API_KEY)
```

```python
# Read the response and set it as dictionay
response = res.json()
# Since all the information is with the key 'data' under 'dataset' it can be subset
zeiss_prices = response['dataset']['data']
zeiss_prices
```

```python
# Construct a daily ordered dict of the data 
data = collections.OrderedDict()
for day in zeiss_prices:
    data[datetime.date.fromisoformat(day[0])] = [day[1], day[2], day[3], day[4], day[6]] # We are interested in columns; 'Open', 'High', 'Low', 'Close' and 'Traded Volume'
```

```python
# 1 - Calculating highest and lowest opening prices for the period
opening_prices = []
for v in data.values():
    opening_prices.append(v[0])
    
highest_opening = max(p for p in opening_prices if p is not None) # Since there are null values max and min would return error if they're not excluded
lowest_opening = min(p for p in opening_prices if p is not None)

print(f'Highest opening price was: {highest_opening} and lowest was: {lowest_opening}')
```

```python
# 2 - Calculating the largest change in one day (based on highest and lowest)
daily_changes = []
for v in data.values():
    try:
        daily_changes.append(abs(v[0] - v[1]))
    except TypeError:
        daily_changes.append(None)
print(f'Highest daily change was: {max(p for p in daily_changes if p is not None)}')
```

```python
# 3 - Calculating largest change between any two days (based on close)
change_for_two_days = [0] # First day won't have a change
for k in data.keys():
    delta = datetime.timedelta(days=1)
    try:
        change_for_two_days.append(abs(data[k][3] - data[k - delta][3]))
    except KeyError:
        change_for_two_days.append(None)
        
largest_change = (max(p for p in change_for_two_days if p is not None))
print(f'Largest change between two days was: {largest_change}')
```

```python
# 4 - Calculating average daily trading volume
trading_volumes = []
for v in data.values():
    trading_volumes.append(v[4])
print(f'Average daily trading volume was: {sum(trading_volumes)/len(trading_volumes)}')
```

```python
# 5 - Calculating median trading volume
def median(lst):
    n = len(lst)
    if len(lst) % 2 == 0 : 
        return (lst[n//2 - 1] + lst[n//2]) / 2
    else: 
        return lst[n//2]

print(f'Median daily trading volume was: {median(trading_volumes)}')
```
